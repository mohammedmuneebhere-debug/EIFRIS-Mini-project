```python
# 3_prediction_interface.ipynb

import pickle
import numpy as np
```

```python
# Load model
with open('//content/drive/MyDrive/Colab Notebooks/mini/disaster_model.pkl', 'rb') as f:
    model = pickle.load(f)
```

```python
# Sample input: [temperature, rainfall, wind_speed, seismic_activity, ...]
user_input = np.array([[24.5, 80.0, 85.0, 0.3, 100.0, 1011.0, 7.0, 1009.0, 8.0]])  # Replace with actual features
```

```python
# Predict
prediction = model.predict(user_input)[0]
```

```python
# Decode label
labels = ['Flood', 'Earthquake', 'Storm', 'Landslide', 'Tsunami', 'Volcanic Eruption', 'Blizzard']
print(f"Predicted Disaster: {labels[prediction]}")
```
