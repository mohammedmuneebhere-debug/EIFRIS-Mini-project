```python
# Simulate map as a graph
G = nx.Graph()
```

```python
# üî∏ Define roads and danger levels (lower = safer)
G.add_edge("User_Location", "Bridge_A", weight=3)  # medium risk
G.add_edge("User_Location", "Road_1", weight=1)    # low risk
G.add_edge("Road_1", "Shelter_1", weight=2)
G.add_edge("Road_1", "Flooded_Street", weight=9)   # high risk
G.add_edge("Flooded_Street", "Shelter_2", weight=8)
G.add_edge("Bridge_A", "Shelter_1", weight=4)
G.add_edge("Bridge_A", "Shelter_2", weight=5)
```

```python
# üîπ Compute the safest path (lowest weighted risk)
safest_path = nx.dijkstra_path(G, source="User_Location", target="Shelter_1", weight='weight')
safest_risk_score = nx.dijkstra_path_length(G, source="User_Location", target="Shelter_1", weight='weight')
```

```python
# üî∏ Output results
print("‚úÖ Safest Path:", " ‚ûú ".join(safest_path))
print("‚ö†Ô∏è  Total Risk Score:", safest_risk_score)
```
