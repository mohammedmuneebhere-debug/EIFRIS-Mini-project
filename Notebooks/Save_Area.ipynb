```python
%pip install geopy folium
from geopy.geocoders import Nominatim
import folium
```

```python
# Initialize geocoder
geolocator = Nominatim(user_agent="eifris_app")
```

```python
# üîπ Get location name (can be dynamic from user)
user_place = "Hyderabad"  # replace this with user input later
```

```python
# üîπ Get latitude & longitude
location = geolocator.geocode(user_place)
```

```python
if location:
    print("üìç Location:", location.address)
    print("üåê Coordinates:", (location.latitude, location.longitude))
else:
    print("‚ùå Could not find the location")
```

```python
# Use fetched coordinates
user_lat, user_lon = location.latitude, location.longitude
```

```python
# üî∏ Sample shelters (in same city - can be dynamically fetched later)
shelters = {
    "Shelter A": (user_lat + 0.05, user_lon + 0.05),
    "Shelter B": (user_lat - 0.03, user_lon + 0.07),
    "Shelter C": (user_lat + 0.04, user_lon - 0.04)
}
```

```python
# üîπ Create a map
m = folium.Map(location=[user_lat, user_lon], zoom_start=12)

```

```python
# üîπ Add user marker
folium.Marker(
    [user_lat, user_lon],
    tooltip="You Are Here",
    icon=folium.Icon(color='blue')
).add_to(m)
```

```python
# üîπ Add shelter markers
for name, coords in shelters.items():
    folium.Marker(
        location=coords,
        tooltip=name,
        icon=folium.Icon(color='green', icon='home')
    ).add_to(m)
```
