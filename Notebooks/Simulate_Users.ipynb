```python
# üßç Sample users with name, location, and role
users = [
    {"name": "Rahul", "age": 70, "lat": 12.96, "lon": 77.6, "role": "elderly"},
    {"name": "AIIMS Hospital", "lat": 28.6, "lon": 77.2, "role": "hospital"},
    {"name": "Riya", "age": 28, "lat": 18.5, "lon": 73.9, "role": "civilian"},
    {"name": "Bridge Control Center", "lat": 12.94, "lon": 77.58, "role": "infrastructure"}
]
```

```python
# üó∫Ô∏è Define geo-fence (example flood zone around Bangalore)
# Format: [lat_min, lat_max, lon_min, lon_max]
flood_zone = [12.90, 13.00, 77.55, 77.65]
```

```python
# üß† Function to check if user is inside the flood zone
def is_inside_zone(user_lat, user_lon, zone_bounds):
    lat_min, lat_max, lon_min, lon_max = zone_bounds
    return lat_min <= user_lat <= lat_max and lon_min <= user_lon <= lon_max
```

```python
# üö® Detect users inside the flood zone
print("üìç Users in danger zone:")
for user in users:
    if is_inside_zone(user["lat"], user["lon"], flood_zone):
        priority = user["role"]
        print(f"üö® Alert: {user['name']} ({priority.upper()}) is inside the geo-fenced danger zone.")
```
