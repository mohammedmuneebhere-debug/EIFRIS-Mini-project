```python
import folium
from folium import plugins

# Same user data and flood zone
users = [
    {"name": "Rahul", "lat": 12.96, "lon": 77.6, "role": "elderly"},
    {"name": "AIIMS Hospital", "lat": 28.6, "lon": 77.2, "role": "hospital"},
    {"name": "Riya", "lat": 18.5, "lon": 73.9, "role": "civilian"},
    {"name": "Bridge Control Center", "lat": 12.94, "lon": 77.58, "role": "infrastructure"}
]
```

```python
# Define flood zone near Bangalore
flood_zone = [12.90, 13.00, 77.55, 77.65]

```

```python
# Create map centered around Bangalore
m = folium.Map(location=[12.95, 77.6], zoom_start=10)
```

```python
# ğŸ”º Priority levels (lower = higher priority)
priority_order = {
    "elderly": 1,
    "hospital": 2,
    "infrastructure": 3,
    "civilian": 4
}
```

```python
# Add flood zone as a rectangle (geo-fence)
folium.Rectangle(
    bounds=[[flood_zone[0], flood_zone[2]], [flood_zone[1], flood_zone[3]]],
    color="red",
    fill=True,
    fill_opacity=0.3,
    tooltip="Danger Zone"
).add_to(m)
```

```python
# Function to check if user is inside zone
def is_inside_zone(user_lat, user_lon, zone_bounds):
    lat_min, lat_max, lon_min, lon_max = zone_bounds
    return lat_min <= user_lat <= lat_max and lon_min <= user_lon <= lon_max
```

```python
# Role to icon color
role_colors = {
    "elderly": "purple",
    "hospital": "green",
    "civilian": "blue",
    "infrastructure": "orange"
}
```

```python
# Add user markers
for user in users:
    inside = is_inside_zone(user["lat"], user["lon"], flood_zone)
    icon_color = "red" if inside else role_colors.get(user["role"], "gray")
    folium.Marker(
        location=(user["lat"], user["lon"]),
        tooltip=f"{user['name']} ({user['role']})",
        icon=folium.Icon(color=icon_color)
    ).add_to(m)

# Show map
m
```

```python
# ğŸ” Geo-fence check
def is_inside_zone(user_lat, user_lon, zone_bounds):
    lat_min, lat_max, lon_min, lon_max = zone_bounds
    return lat_min <= user_lat <= lat_max and lon_min <= user_lon <= lon_max
```

```python
# ğŸ“¨ Simulate alert system
alerts = []
for user in users:
    if is_inside_zone(user["lat"], user["lon"], flood_zone):
        role = user["role"]
        message = f"ğŸš¨ URGENT: Dear {user['name']}, you are in a DANGER ZONE. Please evacuate immediately to the nearest shelter."
        if role == "elderly":
            message += " Assistance is being dispatched for you."
        elif role == "hospital":
            message += " Please prepare emergency units and patient evacuation."
        elif role == "infrastructure":
            message += " Secure critical operations and staff."

        alerts.append({
            "name": user["name"],
            "role": role,
            "priority": priority_order[role],
            "message": message
        })

# ğŸ“‹ Sort alerts by priority
alerts = sorted(alerts, key=lambda x: x["priority"])
```

```python
# ğŸ–¨ Display alerts in order
print("ğŸ“² Sending Emergency Alerts:")
for alert in alerts:
    print(f"\nTo: {alert['name']} ({alert['role'].upper()})")
    print(f"Message: {alert['message']}")
```
