```python
# 1_data_preprocessing.ipynb

import pandas as pd

# Load your dataset (or use a public one)
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mini/historical_disaster_data.csv')
df
```

```python
# Basic EDA
print(df.head())
print(df.info())
```

```python
# Drop missing values
df.dropna(inplace=True)
```

```python
# Encode target labels (e.g., Flood=0, Earthquake=1, etc.)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['disaster_type'] = le.fit_transform(df['disaster_type'])
```

```python
# Save preprocessed data
df.to_csv('//content/drive/MyDrive/Colab Notebooks/mini/cleaned_data.csv', index=False)
```
